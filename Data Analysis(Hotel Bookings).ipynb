```python
df["Month"] = df['reservation_status_date'].dt.month

plt.figure(figsize = (16,8))
sns.countplot(x = "Month", data = df, hue = "is_canceled")
plt.title("Reservation Status Per Month", fontsize = 20)
plt.xlabel("month", fontsize = 20)
plt.ylabel("number of reservation", fontsize = 20)
plt.legend(['not canceled', 'canceled'])
plt.savefig("Reservation Status Per Month.png", dpi = 200)
plt.show()
```

**From above plot we can conclude that August exhibits the highest number of non-cancellations, while January shows the least**

**and August exhibits the lowest number of cancellations, while January shows the highest**

```python
#df["Month"].unique()
# we have 1 to 12 months in our column Month
```

```python
monthly_adr_sum = df[df['is_canceled'] == 1].groupby('Month')[['adr']].sum().reset_index()
print(monthly_adr_sum)

plt.figure(figsize = (15,8))
plt.title("ADR per month")
sns.barplot(x= 'Month', y = 'adr', data = monthly_adr_sum, palette = "husl")
plt.xlabel("Month", fontsize = 20)
plt.ylabel("adr", fontsize = 20)
plt.savefig("ADR per month.png", dpi = 200)
plt.show()
```

**The plot indicates that, on average, prices were lower in August, leading to a higher number of non-cancellations. Guests may have found the lower prices attractive, resulting in increased bookings and fewer cancellations during that period.**

**The plot provides evidence supporting our hypothesis that higher prices are associated with increased cancellations. The observed trend indicates that as the Average Daily Rate (ADR), and consequently prices, rise, there is a corresponding increase in cancellations. This insight reinforces our understanding of the relationship between pricing and cancellation behavior**

**This clear correlation highlights the impact of pricing on guest behavior, affirming that higher prices tend to coincide with a higher likelihood of cancellations. It is plausible that guests may opt to cancel reservations when confronted with elevated prices, underscoring the role of pricing strategies in shaping reservation outcomes**

```python
cancelled_df_adr = cancelled_data.groupby("reservation_status_date")[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values("reservation_status_date", inplace = True)


not_cancelled_data = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby("reservation_status_date")[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values("reservation_status_date", inplace = True)

plt.figure(figsize = (20,9))
plt.title("Average Daily Rate", fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = "not cancelled")
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = "cancelled")
plt.xlabel('Reservation Status Date', fontsize = 20)
plt.ylabel('Average Daily Rate (ADR)', fontsize = 20)
plt.legend(fontsize = 15)
plt.show()

# we can use this formate also
# plt.plot('reservation_status_date', 'adr', data = not_cancelled_df_adr, label = "not cancelled")
#plt.plot('reservation_status_date', 'adr', data = cancelled_df_adr, label = "cancelled")
```

**from above we can see that datset is inconsistent before 2016, very less data. so now we will take data only from 2016 to 2017-09 only.**

**from above plot as ADR is higher cancellation rates are higher.**

**spikes in plot are only on weekends or holidays**

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>"2016") & (cancelled_df_adr['reservation_status_date']<"2017-09")]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>"2016") & (not_cancelled_df_adr['reservation_status_date']<"2017-09")]
```

```python
plt.figure(figsize = (20,9))
plt.title("Average Daily Rate", fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = "not cancelled")
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = "cancelled")
plt.xlabel('Reservation Status Date', fontsize = 20)
plt.ylabel('Average Daily Rate (ADR)', fontsize = 20)
plt.legend(fontsize = 20)
plt.savefig("Average Daily Rate.png", dpi = 200)
plt.show()
```

**In our analysis, we noticed that data before 2016 was inconsistent, so we decided to focus on the period from 2016 to September 2017 for a more reliable view. What caught our attention in the plot is that when prices (ADR) go up, cancellations tend to rise as well. This suggests a connection between higher prices and more cancellations.**. 

**Something interesting we observed is that spikes in the plot usually happen on weekends or holidays. This hints at a possible influence of timing on reservation patterns.**
